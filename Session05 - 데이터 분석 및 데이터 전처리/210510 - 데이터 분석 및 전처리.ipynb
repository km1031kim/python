## np.where 활용 데이터 조작하기
## 1. np.where ( 복잡한 로직에는 힘듬)
## 2. 함수 활용 (복잡한 로직에 좋음)
## 3. loc 함수 활용 ( 가독성이 뛰어나나 속도가 떨어진다)
## 4. map 함수를 사용할 수도 있다.(머신러닝에서 많이 사용된덩)

### 1. np.where 방식

```python
## 데이터프레임명['컬렴명'] = np.where( 조건식, yes값, no값)
```

```python
customerData["PRODUCTAGE_NEW"] = np.where ( customerData.PRODUCTAGE < 1, 1, 0)
```

```python
## 검증로직
```

```python
customerData.loc[ (customerData.PRODUCTAGE < 1) &
                  (customerData.PRODUCTAGE_NEW != 1)]
```

```python
## 여러번 쓸 수도 있다.
```

```python
customerData["PRODUCTAGE_NEW"] = np.where ( customerData.PRODUCTAGE < 1, 1,
                                           np.where(customerData.PRODUCTAGE < 2, 2, 3))
```

#### 2. 함수활용

```python
# 함수 정의. inputValue < 1 : 1,  
#           1 <= inputValue < 2 : 2 
#           inputValue >= 2 : 3
```

```python
def valueClustering(inValue):

    # inValue = 2.5

    outValue = inValue

    if inValue < 1:
        inValue = 1
    elif inValue < 2:
        outValue = 2
    else: 
        outValue = 3

    return(outValue)

```

```python
## 컬럼명.apply(함수이름)
```

```python
customerData["PRODUCTAGE_NEW"] = customerData.PRODUCTAGE.apply(valueClustering)
```

```python
customerData
```

#### 3. loc 방식(비추).. 독립적으로 따로 놀기에 앞 조건을 따지지 않는다.

```python
# customerData.loc[ 조건식, "컬럼"] = 할당 값
```

```python
customerData.loc[ customerData.PRODUCTAGE < 1, "PRODUCTAGE_NEW"] = 1
```

```python
customerData.loc[(customerData.PRODUCTAGE < 1) &
                 (customerData.PRODUCTAGE_NEW != 1)]
```

#### 이런식으로 좀 귀찮아진다

```python
customerData.loc[ (customerData.PRODUCTAGE >= 1) & (customerData.PRODUCTAGE < 2), "PRODUCTAGE_NEW"] = 2
```

```python
customerData.loc[ ( customerData.PRODUCTAGE >= 2), "PRODUCTAGE_NEW"] = 3
```

```python
customerData.loc[ (customerData.PRODUCTAGE < 1) & (customerData.PRODUCTAGE_NEW != 1)]
```
