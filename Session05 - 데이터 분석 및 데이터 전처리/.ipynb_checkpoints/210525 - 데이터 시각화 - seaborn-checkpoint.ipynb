### * 차트의 종류

#### 단일변수 차트 countplot - 카운트차트(변수의 빈도수 분포), distplot - 히스토그램(연속형 변수의 빈도수 분포)

#### 두 변수 이상 차트 barplot, boxplot, lineplot, scatterplot, jointplot, heatmap

```python
csData.columns
```

#### COUNTPLOT

```python
### sns는 groupby 가능(hue)
sns.countplot(data = csData,
              x="EMI", hue = "CUSTTYPE" )
```

#### DISTPLOT

```python
figure = plt.figure(figsize=(18,6)) ## 도화지그리지
ax1 = plt.subplot(1,1,1)
sns.distplot(csData.AVGPRICE, hist = True)
```

### bar plot - 하나의 컬럼을 기준으로 다른 변수의 평균을 그래프로 표현

```python
sns.barplot(x = "EMI", y = "AVGPRICE", data = csData, hue = "CUSTTYPE", ci = "sd")
```

### box plot - 정해진 기준에 따라 변수의 분포를 시각화. 어떤 데이터에 대해 아웃라이어가 있는지 확인

```python
### 4분위수 (4등분하는 위치의 수)
```

```python
sns.boxplot(data=csData.loc[csData.EMI != 1], x="EMI", y="AVGPRICE")
```

###### 맨 위쪽 가로가 3사분위수
###### 박스의 맨 위 =  (중간값이 없는 위쪽 부분의 중간. 3사분위수의 75%)
###### 박스의 중간선이 중앙값(평균과는 다른 값)
###### 박스의 맨 밑줄 =  (중간값이 없는 곳의 중간. 1사분위수의 25%)
###### IQR = 3사분위수 - 1사분위수
###### 맨 아래쪽 가로가 1사분위수
###### 범위 밖의 점이 이상치
###### 이상치가 없는 경우  
###### - 1사분위수 - (IQR * 1.5) 가 데이터의 최소값 보다 작으면 최소값을 MINUMUM 값으로 대체
###### - 3사분위수 + (IQR * 1.5) 가 데이터의 최대값 보다 크면 최대값을 MAXIMUM 값으로 대체

### LINEPLOT -  x축과 y축에 값을 할당하고 x축에 따른 y축 변수의 변화량을 시각화

```python
import pandas as pd

featuresData = pd.read_csv("../dataset/feature_regression_example.csv")

sns.lineplot(data=featuresData.loc[featuresData.YEAR==2015], x = "YEARWEEK", y = "QTY")
```
