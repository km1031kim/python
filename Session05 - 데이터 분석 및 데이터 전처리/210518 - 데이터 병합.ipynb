## 3. 데이터 병합

```python
joinMaster.loc[4, "AP2ID"] = "A05"

joinMaster.loc[4, "REGIONNAME"] = "인도"
```

```python
joinMaster
```

```python
joinSellout
```

```python
joinSellout.loc[4, "REGIONID"] = "A06"

joinSellout.loc[4, "PRODUCT"] = "ST005"
```

```python
pd.merge(left=joinSellout,
         right=joinMaster,
         on="REGIONID",
         how="left") 
```

## 조인키 대상은 중복을 허락하지 않는 기본키여야 한다. 조인하는키가 기본키가 아니면, 데이터가 늘어난다

```python
testMaster = joinMaster.loc[0:1]
```

#### testmaster의 regionid가 a02인것을 찾아서 해당 컬럼의 regionid를 a01로 바꾼다.

```python
testMaster.loc[testMaster.REGIONID=="A02", "REGIONID"] = "A01"
```

```python
joinSellout
```

```python
testMaster
```

#### 아래와 같이 데이터의 중복이 일어남

```python
pd.merge(left=joinSellout,
         right=testMaster,
         on="REGIONID",
         how="inner")
```

```python
joinSellout.rename(columns={"REGIONID":"AP2ID"},inplace=True)
```

```python
joinMaster
```

```python
joinSellout
```

```python
pd.merge(left = joinSellout,
         right = joinMaster,
         left_on = "REGIONID",
         right_on = "AP2ID",
         how="inner")
```

## 프로모션 정보(연습)

```python
promotionCloumns = ["SUBSIDIARY", "CARD_NAME", "PROMOTION_NAME", "PROMOTION_RATE"]
```
