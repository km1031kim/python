## 데이터 저장(postgres에 저장)

```python
import psycopg2 
from sqlalchemy import create_engine
```

```python
mylocalpgUrl = "postgresql://postgres:postgres@127.0.0.1/postgres"
```

```python
engine_pg = create_engine(mylocalpgUrl)
```

## 컬럼이름, 테이블이름(메타정보) 들은 소문자로! (포스트그레인 경우에)

```python
resultTableName = "pg_out"
```

```python
data.to_sql(name=resultTableName,
            con=mylocalpgUrl, if_exists="replace", index=False
            data.columns = ["regionid", "productgroup", "yearweek", "volume"])
```

```python
targetTable = "pg_out"
```

```python
## 저장한 데이터 불러오기
sqlReadData = pd.read_sql(sql="select * from pg_out", con=engine_pg)
```

```python
pd.read_sql_table(table_name = "pg_out", con=engine_pg)
```

## 실습해보기 - 방화벽 뚫어야함.

## 1. port 열기, 방화벽 -> 고급 설정 -> 인바운드 규칙(내 피씨에 서버를 설정한 경우) -> 새 규칙 -> 포트 허용
## 2. 데이터베이스 ip접근 관리(ip획득 후) -> 설치폴더이동 -> 10 -> data -> pg_hba.conf -> 메모장으로 열고 맨 마지막줄에 host all all 0.0.0.0/0 md5 입력(전부 다 접근 허용. md5는 암호화 모듈)
## 3. 서비스 창 실행 -> postgersql 우클릭 후 다시시작.

### 혁중이형 아이피로 접근, 데이터 넣기

```python
jungip = "postgresql://postgres:1234@192.168.110.7/postgres"   ## 아이디:비밀번호@ip

## 보낼 데이터
targetData = "../dataset/kopo_product_volume.csv"  ## 파일 이름
data = pd.read_csv(targetData, encoding="ms949")  ## csv파일로 ms949 인코딩
data.columns = ["regionid", "productgroup", "yearweek", "volume"] ## 데이터 컬럼 변경해서 보내기(소문자로)

## 엔진 설정
engine_pg = create_engine(jungip)

## 상대방 아이피 데이터베이스에 저장될 데이터 파일 이름 설정
resultTableName = "kopo_product_volume4"

data.to_sql(name=resultTableName, ## sql보내기
            con=jungip, if_exists="replace", index=False)
```

```python
targetTable = "pg_out"
```

```python
data.rename(columns={"REGIONID":"SID",
                     "YEARWEEK":"SNAME"} )
```
